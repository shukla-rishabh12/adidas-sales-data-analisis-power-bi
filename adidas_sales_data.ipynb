```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel("C:/Users/dell/Downloads/Adidas US Sales Datasets.xlsx")
df.head(5)

```

```python
df=pd.read_excel("C:/Users/dell/Downloads/Adidas US Sales Datasets.xlsx",skiprows=4)
df
```

```python
df.columns
df.info()
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(5)
df.describe()
```

```python
sales_by_region= df.groupby('Product')['Total Sales'].sum()
print(sales_by_region)
plt.figure(figsize=(10,3))
plt.bar(sales_by_region.index,sales_by_region.values,color='yellow')
plt.xlabel('product')
plt.ylabel('total sales(in ten-carors)')
plt.xticks(rotation=85)
plt.show()
```

```python
sales_mathod=df.groupby('Sales Method')['Total Sales'].sum()
plt.pie(sales_mathod.values,labels=sales_mathod.index,autopct='%1.1f%%',colors=['green','yellow','blue','orange'])
plt.show()
```

```python
df['month']=df['Invoice Date'].dt.strftime('%B')
df['year']=df['Invoice Date'].dt.strftime('%Y')
df.head(5)
```

```python
sales_by_month=df.groupby('month')['Total Sales'].sum()
print(sales_by_month)

plt.plot(sales_by_month.index,sales_by_month.values,color='red',marker='o')
plt.xlabel('month names')
plt.title('slaes by months')
plt.grid(True)
plt.ylabel('total sales in carores')
plt.xticks(rotation=70)
plt.show()
```

```python
sales_by_year=df.groupby('year')['Total Sales'].sum()
print(sales_by_year)

plt.barh(sales_by_year.index,sales_by_year.values,color='green')
plt.ylabel('year names')
plt.title('slaes by myear')
plt.grid(True)
plt.xlabel('total sales in en-carores')
plt.xticks(rotation=70)
plt.show()
```
